```python
date_from="2023-03-12"
date_to="2023-03-14"
parameter="2"
limit="1000"
page="1"
spatial="location"
temporal="day"
```

```python
#check for thailand
country_list="th"
```

```python
#create the url to be used 
url=f"https://api.openaq.org/v2/averages?date_from={date_from}&date_to={date_to}&parameter={parameter}&country_id={country_list}&limit={limit}&page={page}&offset=0&sort=desc&spatial={spatial}&temporal={temporal}&group=false"
```

```python
#from doc: 
#https://api.openaq.org/docs#/v2/averages_v2_get_v2_averages_get
```

```python
# get the json and check the length
headers = {"accept": "application/json"}
response = requests.get(url, headers=headers)
json_response=response.json()
len(json_response['results'])
```

```python
json_response
```

```python
#transform it in a table
pd_response=json_normalize(json_response['results']) 
pd_response=pd_response[['id','day','name','average','measurement_count','parameter','unit']]
pd_response['country']=country_list
pd_response
```

```python
#remove every data point over 1000 as it's likel an error
threshold=1000
pd_response=pd_response[pd_response['average']<=threshold]
pd_response=pd_response[pd_response['average']>0]
```

```python
#change the index to the day
pd_response=pd_response.set_index('day')
```

```python
#plot the graph
pd_response['average'].plot()
```

# Get the location informations by country

```python
# get the location geolocation 
limit_location='5000'
page_location='1'
country_list_location=['th']

country_list_location_str=["&country_id="+country for country in country_list_location]
country_list_location_str="".join(country_list_location_str)
country_list_location_str

url_location=f"https://api.openaq.org/v2/locations?limit={limit_location}&page={page_location}&offset=0&sort=desc&radius=1000{country_list_location_str}&order_by=lastUpdated&dumpRaw=false"
```

```python
url_location
```

```python
headers = {"accept": "application/json"}
response_location = requests.get(url_location, headers=headers)
json_response_location=response_location.json()
```

```python
json_response_location['results']
```

```python
df_locations=json_normalize(json_response_location['results']) [['id','city','name','entity','country','isMobile','sensorType','coordinates.latitude','coordinates.longitude','coordinates']]
```
